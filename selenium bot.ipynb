```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import Select
import time
import pandas as pd
```

```python
website='https://www.adamchoi.co.uk/overs/detailed'

path=r'"C:\Users\subha\OneDrive\Desktop\chromedriver.exe"'


driver=webdriver.Chrome(path)

driver.get(website)
```

```python
all_matches_button=driver.find_element(By.XPATH, '//label[@analytics-event="All matches"]')
all_matches_button.click()
```

```python
dropdown = Select(driver.find_element(By.ID, "country"))
dropdown.select_by_visible_text('Spain')


time.sleep(3)
```

```python
matches= driver.find_elements(By.TAG_NAME, 'tr')


for match in matches:
    print(match.text)
```

```python
date=[]
home_team=[]
score=[]
away_team=[]

for match in matches:                                             #|//tr|/td[]
    date.append(match.find_element(By.XPATH, './td[1]').text)     #|  . |/td[] 
    home_team.append(match.find_element(By.XPATH, './td[2]').text)
    score.append(match.find_element(By.XPATH, './td[3]').text)
    away_team.append(match.find_element(By.XPATH, './td[4]').text) 
```

```python
data = {'date': date, 'home_team':home_team, 'score':score,'away_team':away_team}
data
```

```python
df = pd.DataFrame(data)
df.to_csv('spain_football_data.csv',index=False)
df
```

```python
driver.quit()
```
